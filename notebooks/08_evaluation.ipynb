```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
baseline_results = pd.read_csv("../results/baseline_results.csv")
rnn_results = pd.read_csv("../results/rnn_results.csv")
lstm_results = pd.read_csv("../results/lstm_results.csv")
gru_results = pd.read_csv("../results/gru_results.csv")
```

```python
all_results = pd.concat(
    [baseline_results,rnn_results,lstm_results,gru_results],
    ignore_index = True
)
all_results
```

```python
all_results.to_csv("../results/all_model_results.csv", index=False)
print("All model results saved.")
```

```python
all_results.sort_values(by="Validation F1",ascending=False)
```

```python
all_results.sort_values(by="Test F1",ascending = False)
```

```python
plt.figure(figsize=(8, 5))
plt.bar(all_results["Model"], all_results["Test Accuracy"])
plt.ylabel("Test Accuracy")
plt.title("Model Comparison - Test Accuracy")
plt.xticks(rotation=15)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(all_results["Model"], all_results["Test F1"])
plt.ylabel("Test F1 Score")
plt.title("Model Comparison - Test F1 Score")
plt.xticks(rotation=15)
plt.show()
```

```python
comparison_df = all_results[["Model", "Validation Accuracy", "Test Accuracy"]]
comparison_df
```

```python
comparison_df.plot(
    x="Model",
    kind="bar",
    figsize=(9, 5)
)

plt.ylabel("Accuracy")
plt.title("Validation vs Test Accuracy")
plt.xticks(rotation=15)
plt.show()
```

```python
comparison_f1_df = all_results[["Model", "Validation F1", "Test F1"]]
comparison_f1_df
```

```python
comparison_f1_df.plot(
    x="Model",
    kind="bar",
    figsize=(9, 5)
)

plt.ylabel("F1 Score")
plt.title("Validation vs Test F1 Score")
plt.xticks(rotation=15)
plt.show()
```

```python
best_model_row = all_results.sort_values(by="Test F1", ascending=False).iloc[0]
best_model_row
```

```python
print("Best model based on Test F1:")
print(f"Model: {best_model_row['Model']}")
print(f"Test Accuracy: {best_model_row['Test Accuracy']:.4f}")
print(f"Test Precision: {best_model_row['Test Precision']:.4f}")
print(f"Test Recall: {best_model_row['Test Recall']:.4f}")
print(f"Test F1: {best_model_row['Test F1']:.4f}")
```
