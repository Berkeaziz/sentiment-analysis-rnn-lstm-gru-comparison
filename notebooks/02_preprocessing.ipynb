```python
def clean_text(text:str) -> str:
    text = str(text).lower()

    text= re.sub(r"<.*?>"," ",text)

    text=re.sub(r"http\S+|www\S+"," ",text)

    text = re.sub(r"[^a-zA-Z\s]"," ",text)

    text = re.sub(r"\s+"," ", text).strip()

    return text
```

```python
NEGATION_WORDS ={"not","no","never","nor"}
```

```python
def spacy_preprocess(text: str) -> str:
    doc = nlp(text)
    tokens = []
    
    for token in doc:
        if token.is_space or token.is_punct:
            continue
        
        # stopword ise çıkar ama negation ise koru
        if token.is_stop and token.text not in NEGATION_WORDS:
            continue
        
        lemma = token.lemma_.strip()
        
        if not lemma:
            continue
        
        tokens.append(lemma)
    
    return " ".join(tokens)
```
